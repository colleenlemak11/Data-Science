# Google Trends Regression

```python
import pandas as pd

# read and store data in csv file
df = pd.read_csv("NCAA_basketball.csv")

x = df["NCAA basketball championship: (United States)"]
x = pd.Series(x)
x.replace("<1", 0, inplace=True)
x = x.astype(int)

y = df["Gonzaga University: (United States)"]
y = pd.Series(y)
y.replace("<1", 0, inplace=True)
y = y.astype(int)  
```

```python
import matplotlib.pyplot as plt

# plotting the scatter plot
plt.scatter(x, y)
plt.show()
```

```python
import scipy.stats as stats
import numpy as np

# calculating statistics
m, b, r, p_val, std_err = stats.linregress(x, y)
print("m:", m, "b:", b, "r:", r)
print("Our r value is greater than 0.5! :)")
```

```python
# plotting the scatter plot with line of best fit
plt.scatter(x, y)

x_min = x.min()
x_max = x.max()

y_min = m * x_min + b
y_max = m * x_max + b

plt.plot([x_min, x_max], [y_min, y_max], c="green", lw=5, label="line of best fit")
plt.xlabel("NCAA basketball championship")
plt.ylabel("Gonzaga University")
plt.title("Gonzaga Search Term Statistics")
plt.legend()
plt.show()
```
